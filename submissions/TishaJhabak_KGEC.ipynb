## Spliting the data.

```python
X=king_data.copy()
y=X.pop('Column37')

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.7,train_size=0.3)
```

```python
y_test.head(10)
```

## Training and predicting the model using KNN algo.



```python
from sklearn.preprocessing import OneHotEncoder
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

one_hot_train=pd.DataFrame(OH_encoder.fit_transform(X_train[object_col]))
one_hot_test=pd.DataFrame(OH_encoder.transform(X_test[object_col]))
one_hot_train.index = X_train.index
one_hot_test.index = X_test.index
```

```python
num_X_train = X_train.drop(object_col, axis=1)
num_X_valid = X_test.drop(object_col, axis=1)
```

```python
OH_X_train = pd.concat([num_X_train, one_hot_train], axis=1)
OH_X_valid = pd.concat([num_X_valid, one_hot_test], axis=1)
```

```python
y_train.unique()
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn

model=knn(n_neighbors=1,
          leaf_size=1,
          algorithm='brute',
          weights='distance',
          p=1
          )

model.fit(OH_X_train,y_train)
```

```python
pred=model.predict(OH_X_valid)
```

## Checking model accuracy.


```python
from sklearn.metrics import accuracy_score

x=accuracy_score(y_test,pred)
print(f'the accuracy score is {x}')
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cm= confusion_matrix(y_test,pred)
```

```python
print(cm)
```

# Trying Other Classifiction Algorithm :
- Decision Tree 
- Random Forest
- Naive Bayes
- Support Vector Machine

```python
from sklearn.model_selection import GridSearchCV    # for getting the best hyperparameters
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
model_params  = {    
    "svm" : {
        "model":SVC(gamma="auto"),
        "params":{
            'C' : [1,10,20],
            'kernel':["rbf"]
        }
    },
    
    "decision_tree":{
        "model": DecisionTreeClassifier(random_state=0),
        "params":{
            'criterion':["entropy","gini"],
            "max_depth":[5,8,9]
        }
    },
    
    "random_forest":{
        "model": RandomForestClassifier(random_state=0),
        "params":{
            "n_estimators":[1,5,10],
            "max_depth":[5,8,9,10]
        }
    },
    "naive_bayes":{
        "model": GaussianNB(),
        "params":{}
    }
}
```

```python
score=[]
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp["model"],mp["params"],cv=4,return_train_score=False)
    clf.fit(OH_X_train,y_train)
    score.append({
        "Model" : model_name,
        "Best_Score": clf.best_score_,
        "Best_Params": clf.best_params_
    })
```

```python
result = pd.DataFrame(score,columns=["Model","Best_Score","Best_Params"])
```

```python
result
```

`In above cases, we get the decision tree to give the best_score on the training set, so we will try using cross validation on a decision tree algorithm as below:`

```python
from sklearn.model_selection import cross_val_score    # cross validating 

clf_dt = DecisionTreeClassifier(max_depth= 9, criterion= 'entropy', random_state=1)
scores = cross_val_score(clf_dt,OH_X_train,y_train,cv=8,scoring="accuracy")
```

```python
scores.mean()*100
```

```python
clf_dt.fit(OH_X_train,y_train)
y_pred = clf_dt.predict(OH_X_valid)
```
